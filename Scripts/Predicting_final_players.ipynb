```python
# Calculate Batsmen and Bowlers Rating for Indian players and Australian players based on key metrics using formula from a conference paper
# formula for batsman  = 0.4262*Ave + 0.2566*Inns + 0.1510*SR + 0.0787*Centuries + 0.0556*Fifties - 0.0328*Zeros
# formula for bowler = 0.3269*Overs + 0.2846*Inns + 0.1877*SR + 0.1210*Ave

df_ind_batsmen['Batsmen_Rating'] = 0.4262 * df_ind_batsmen['Ave'] + 0.2566 * df_ind_batsmen['Inns'] + 0.1510 * df_ind_batsmen['SR'] + 0.0787 * df_ind_batsmen['100'] + 0.0556 * df_ind_batsmen['50'] - 0.0328 * df_ind_batsmen['0']
df_ind_bowlers['Bowlers_Rating'] = 0.3269 * df_ind_bowlers['Overs'] + 0.2846 * df_ind_bowlers['Inns'] +  0.1877 * df_ind_bowlers['SR'] + 0.1210 * df_ind_bowlers['Ave']
df_aus_batsmen['Batsmen_Rating'] = 0.4262 * df_aus_batsmen['Ave'] + 0.2566 * df_aus_batsmen['Inns'] + 0.1510 * df_aus_batsmen['SR'] + 0.0787 * df_aus_batsmen['100'] + 0.0556 * df_aus_batsmen['50'] - 0.0328 * df_aus_batsmen['0']
df_aus_bowlers['Bowlers_Rating'] = 0.3269 * df_aus_bowlers['Overs'] + 0.2846 * df_aus_bowlers['Inns'] +  0.1877 * df_aus_bowlers['SR'] + 0.1210 * df_aus_bowlers['Ave']
```

```python
# Accumulate ratings for players with multiple entries
df_ind_batsmen_agg = df_ind_batsmen.groupby('Player')['Batsmen_Rating'].sum().reset_index()
df_ind_bowlers_agg = df_ind_bowlers.groupby('Player')['Bowlers_Rating'].sum().reset_index()
df_aus_batsmen_agg = df_aus_batsmen.groupby('Player')['Batsmen_Rating'].sum().reset_index()
df_aus_bowlers_agg = df_aus_bowlers.groupby('Player')['Bowlers_Rating'].sum().reset_index()
```

```python
# Select top 6 batsmen and top 5 bowlers for playing 11
playing_eleven_ind_batsmen = df_ind_batsmen_agg.nlargest(6, 'Batsmen_Rating')
playing_eleven_ind_bowlers = df_ind_bowlers_agg.nlargest(5, 'Bowlers_Rating')
playing_eleven_aus_batsmen = df_aus_batsmen_agg.nlargest(6, 'Batsmen_Rating')
playing_eleven_aus_bowlers = df_aus_bowlers_agg.nlargest(5, 'Bowlers_Rating')
```

```python
# Display the selected playing elevens
print("Playing Eleven IND - Batsmen:")
print(playing_eleven_ind_batsmen[['Player', 'Batsmen_Rating']])

print("\nPlaying Eleven IND - Bowlers:")
print(playing_eleven_ind_bowlers[['Player', 'Bowlers_Rating']])
```

```python
# Display the selected playing elevens
print("Playing Eleven AUS - Batsmen:")
print(playing_eleven_aus_batsmen[['Player', 'Batsmen_Rating']])

print("\nPlaying Eleven AUS - Bowlers:")
print(playing_eleven_aus_bowlers[['Player', 'Bowlers_Rating']])
```
