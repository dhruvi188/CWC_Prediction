# Web Scraping

```python
#Importing the libraries
import requests
from bs4 import BeautifulSoup
```

```python
#Collect page link
page = requests.get('https://www.espncricinfo.com/records/tournament/batting-most-runs-career/icc-cricket-world-cup-2023-24-15338')
```

```python
#Create a BeautifulSoup object
soup = BeautifulSoup(page.text,'html.parser')
```

```python
table= soup.find("table")
print(table.prettify())
```

```python
headers= table.find_all("span")
print(headers)
```

```python
data_entries=[]
for i in headers:
  data_entry= i.text
  data_entries.append(data_entry)

print(data_entries)
```

```python
#Seperating the column headers
titles = data_entries[:15]
print(titles)
```

```python
rows = table.find_all("tr")
row_data_list = []  # Initialize an empty list to store the lists of row data

for row in rows[1:]:
    # Extracting text from each <td> element
    columns = row.find_all("td")
    row_data = [column.text.strip() for column in columns]

    # Append the row_data list to the row_data_list
    row_data_list.append(row_data)

# Now row_data_list is a list of lists with the extracted data from each row
row_data_list
```

```python
#Dataframe created from the table data
df = pd.DataFrame(row_data_list, columns=titles)
df.head()
```

```python
# Extract country information from parantheses
df['Country'] = df['Player'].str.extract(r'\((\w+)\)')
```

```python
# Removing span column as the data is only for ongoing world cup
df = df.drop('Span', axis=1)
```

```python
df.shape
```

**DATASET DESCRIPTION**

**Shape**- 145 rows and 15 columns

**Columns**-

1) Player Name

2) No. of Matches played

3) No. of Innings Played

4) No. of times batsman remained NOT OUT

5) Runs Scored

6) High Score

7) Batting Average (Runs Scored / Number of times dismissed)

8) Balls Faced

9) Strike Rate ((Runs Scored / Balls Faced) * 100)

10) No of 100s scored

11) No of 50s scored

12) No of ducks(0s)

13) No of 4s

14) No of 6s

15) Country

PREPROCESSING

```python
#Replacing all '-' in columns with '0's
df['100'] = df['100'].replace('-', 0)
df['50'] = df['50'].replace('-', 0)
df['0'] = df['0'].replace('-', 0)
df
```

```python
#removing * from HS column
df['HS'] = df['HS'].str.replace('*', '', regex=False)
df.head()
```

Selecting only the players with more than 250 runs from the dataframe and removing the others as only these players have a fighting chance to be the top scorer in the tournament.

```python
df['Runs'] = pd.to_numeric(df['Runs'], errors='coerce')
min_runs= 250
new_df = df[df['Runs'] > min_runs]
new_df
```

Now keeping players only from Australia, New Zealand, India, and South Africa as they only have chance to be top scorer.

```python
# List of countries to keep
countries = ['AUS', 'IND', 'NZ', 'SA']

# Filter the DataFrame based on the desired countries
new_df = new_df[new_df['Country'].isin(countries)]
new_df = new_df.reset_index(drop=True)

new_df
```

Only 18 players remain and only players from this list have a fighting chance to be the top run scorer of the ICC World Cup 2023.

Creating a new column FORM with the formula

Form = (0.4262 x Average) + (0.2566 x no. of innings) + (0.1510 x SR) + (0.0787 x Centuries) + (0.0556 x Fifties) + (0.0328 x HS)

This column helps predict the current form the batsman is in, in the tournament.
Formula is taken from a research paper by Kalpdrum Passi and Niravkumar Pandey. They have taken the form with stats over a year, instead I have taken the stats ranging only the tournament.

```python
new_df['Ave'] = pd.to_numeric(new_df['Ave'], errors='coerce')
new_df['Inns'] = pd.to_numeric(new_df['Inns'], errors='coerce')
new_df['SR'] = pd.to_numeric(new_df['SR'], errors='coerce')
new_df['100'] = pd.to_numeric(new_df['100'], errors='coerce')
new_df['50'] = pd.to_numeric(new_df['50'], errors='coerce')
new_df['HS'] = pd.to_numeric(new_df['HS'], errors='coerce')

# Apply the formula
new_df['Form'] = (
    0.4262 * new_df['Ave'] +
    0.2566 * new_df['Inns'] +
    0.1510 * new_df['SR'] +
    0.0787 * new_df['100'] +
    0.0556 * new_df['50'] +
    0.0328 * new_df['HS']
)
```
